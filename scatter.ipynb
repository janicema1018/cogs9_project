```
import numpy as np
import matplotlib.pyplot as plt
```

```
line_number = [1133, 1379, 1265, 1392, 1042, 1005, 1107, 1104, 1115, 1123]
score = [83, 91, 91, 90, 89, 82, 85, 72, 69, 75]
```

```
plt.figure(figsize=(8, 8))
plt.scatter(line_number, score, alpha=0.5)
plt.axis([900, 1400, 0, 100])
plt.xlabel("Sheldon's number of lines in each season")
plt.ylabel("score evaluated by audience for each season on Rotten Tomatoes (%)")
plt.title("Relationship between score evaluated by audience \n and Sheldon's number of lines in each season \n of The Big Bang Theory \n")
plt.plot(np.unique(line_number), np.poly1d(np.polyfit(line_number, score, 1))(np.unique(line_number)))
r1 = np.corrcoef(line_number, score)
print(r1[0, 1])
print(r1[1, 0])
```
